```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = open("AngelaNumber.txt", "r")
content_number = file.read()
file.close()
file = open("AngelaColor.txt", "r")
content_color = file.read()
file.close()
```

```python
data_number = np.array(content_number.split("\n")[:-1]).astype(int)
data_color = np.array(content_color.split("\n")[0:10]).astype(int)
len(data_color)
```

```python
x_number = []
for i in range(2, 19):
    x_number.append(i*i)
    
x_color = []
i=0.5
while i < 1.5:
    x_color.append(i)
    i+=0.1
len(x_color)
```

```python
plt.plot(x_number, data_number)
plt.show()
```

```python
plt.plot(x_color, data_color)
plt.show()
```

```python
fig=plt.figure()
ax=fig.add_subplot(111, label="1")
ax2=fig.add_subplot(111, label="2", frame_on=False)

ax.plot(x_number, data_number, color="C0")
ax.set_xlabel("Number of Cubes", color="C0")
ax.set_ylabel("Reaction Time", color="C0")
ax.tick_params(axis='x', colors="C0")
ax.tick_params(axis='y', colors="C0")

ax2.plot(x_color, data_color, color="C1")
ax2.xaxis.tick_top()
ax2.yaxis.tick_right()
ax2.set_xlabel('Color Distance', color="C1") 
ax2.set_ylabel('Reaction Time', color="C1")       
ax2.xaxis.set_label_position('top') 
ax2.yaxis.set_label_position('right') 
ax2.tick_params(axis='x', colors="C1")
ax2.tick_params(axis='y', colors="C1")
```
